```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/2015.csv')
```

Crear Histogramas:

Crea histogramas para las variables Score, GDP per capita, Healthy life expectancy y Freedom to make life choices.

Personaliza los histogramas con colores, bordes y ajustes de transparencia.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Happiness Score'],bins= 5, color='blue', edgecolor='black', alpha=0.7, kde=True)
plt.title("Histograma de los puntajes")
plt.xlabel("Puntaje de felicidad")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True)
plt.show()
```

Distribución del Puntaje de Felicidad:

El histograma muestra que la mayoría de los países tienen un puntaje de felicidad entre 4 y 6, se ppodria decir que la mayoria de los paises tiene una felicidad alta, sin embargo hay algunos pasises que indican una felicidad baja, entre 3 y 2

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Economy (GDP per Capita)'],bins= 5, color='red', edgecolor='black', alpha=0.7, kde=True )
plt.title("Histograma PIB per Capita")
plt.xlabel("PIB per Capita")
plt.ylabel("Frecuencia")
plt.show()
```

El histograma de PIB per capita nos muestra que la mayoria de los pasies se concentra en un rango de 1 en el eje PiB percapita lo que indica que estos paises se encuentran en un rango medio de ingresos, sin embargo el histograma tiene valores mas altos hacia los rangos mas bajos del eje PIP per capita lo que indica que hay mas paises con ingresos mas bajos en este estudio comparado con los paises de mas altos ingresos que se encuentran alrededor de 1.50 del eje PIB per percapita

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Health (Life Expectancy)'],bins= 5, kde=True)
plt.title("Histograma Esperanza de Vida")
plt.xlabel("esperanza de Vida")
plt.ylabel("Frecuencia")
plt.show()
```

El histograma de esperanza de vida indica que la mayoria de los paises se encuentra entre un rango entre 0.6 y 0.8, muy similar tambien al pico que vimos en el histograma de PIB per capita, lo que indica que los paises con mayores ingresos tienen una mayor esperanza de vida, esto lo podriamos relacionar con la cola que va a valores mas altos de esperanza de vida

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Freedom'],bins= 5, color='green', edgecolor='black', alpha=0.7, kde=True)
plt.title("Histograma Libertad")
plt.xlabel("Libertad")
plt.ylabel("Frecuencia")
plt.show()
```

En el histograma de libertad se puede ver que el pico se encuentra cerca de 0.4 lo que indica que la mayoria de los paises tiene un rango medio de libertad, tambien la cola indica que hay un grupo de paises con un mayor nivel de libertad

```python
plt.figure(figsize=(10,6))
sns.histplot(df, x='Happiness Rank', hue='Region', bins=5, color='skyblue', edgecolor='black', alpha=0.7, kde=True)
plt.title("Histograma de Rango de felicidad por region")
plt.xlabel("Rango")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

Crear Boxplots:

Crea boxplots para las mismas variables para visualizar la mediana, el rango intercuartil y los valores atípicos.

Personaliza los boxplots con diferentes colores y estilos. Análisis Combinado:

Combina los histogramas y los boxplots en una sola figura para cada variable.

Interpreta los resultados y escribe un análisis sobre la distribución de cada variable y la presencia de valores atípicos.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Happiness Score', data=df, notch=True, patch_artist=True, boxprops=dict(facecolor='skyblue', color='blue'), medianprops=dict(color='red'), whiskerprops=dict(color='black'), capprops=dict(color='green'))
plt.title("Boxplot de los puntajes")
plt.xlabel("Puntaje de felicidad")
plt.grid(True)
plt.show()
```

El boxplot de felicidad indica que la mayoria de los paises se encuentra en un rango medio  de felicidad, la mediana del boxplot se encuentra un poco mas de 5 lo que indica que la mayoria de los paises tienen un puntaje de felicidad relativamente alto, no se ven valores atipicos o outlier, los bigotes del boxplox son bien largos, y abarcan casi tod el eje de puntaje de felicidad, desde menos de 3 a mas de 7, lo que siguiere que hay paises en el extremos de ambos puntos, algunos con un nivel de bajo puntaje en felicidad y otros con un nivel alto de felicidad

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Economy (GDP per Capita)', data=df, notch=True, patch_artist=True, boxprops=dict(facecolor='black', color='red'), medianprops=dict(color='yellow'), whiskerprops=dict(color='green'), capprops=dict(color='purple'))
plt.title("Boxplot del PIB")
plt.xlabel("Puntaje de PIB")
plt.show()
```

El rango del PIB es relativamente amplio como s emuestra en los bigotes del boxplot , abarcando del 0 hasta 1,75 lo que indica que hay paises con muy bajo PIB y otros con un PIB muy alto. La mediana se encuntra alrededor de 1.00. lo que indica que la mitad de los paises tiene un puntaje menor a 1 del PIB y la otra mitad de paises tiene un puntaje mayor a 1 del PIB.Tampoco se ven valores atipicos o outliers en el boxplot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Health (Life Expectancy)', data=df, color='skyblue')
plt.title("Boxplot de Esperanza de vida")
plt.xlabel("Puntaje de Esperanza de vida")
plt.show()
```

El boxplot de esperanza de vida tiene un rango desde o hasta 1 tal cual como indican los bigotes del boxplot, lo que indica que hay paises con una esperanza de vida muy baja y otros con una muy alta. La mediana de esperanza de vida de los paises se encuentra alrededor de 0,7, esto significa que una gran mayoria de los paises encuestados tiene una esperanza de vida menor a este valor  y hay otros que tienen un rango superior a este valor, como se dijo al principio tambien hay paises con un alto puntaje en esperanza de vida

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Freedom', data=df, notch=False, patch_artist=True, boxprops=dict(facecolor='skyblue', color='green'), medianprops=dict(color='red'), whiskerprops=dict(color='red'), capprops=dict(color='red'))
plt.title("Boxplot de Libertad")
plt.xlabel("Puntaje de Libertad")
plt.show()
```

El boxplot de libertad muestra una concentracion de la caja entre 0.3 y 0.5 lo que indica que hay una gran concentracion de los paises encuestados en este rango medio de libertad. La linea roja que representa ma mediana se encuentra alrededor de los 0.45 lo que indica que la mitad de los paises tienen un puntaje por debajo de este valor y la otra mitad superir a 0.45. tambien los vigotes del boxplot varian desde 0 a 0.7 lo que nuevamente indica que hay paises con nivles muy bajos de libertad y otros muy altos.. tampoco se ven valores atipicos en este boxplot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Happiness Score', y= 'Region', data=df, color='skyblue')
plt.title("Boxplot de Puntuacion de felicidad por region")
plt.xlabel("Puntuacion de felicidad")
plt.ylabel("Region")
plt.show()
```

El boxplot de puntaje de felicidad por region indica que los paises que se encuentran en Europa Occidental, Norteamerica y Australia con puntaje cercano a 8 tiene los mayores indices de felicida, esto se explica por el desarrollo economico, mayores libertades que llevan a tener una mayor esperanza de vida y por lo  tanto pareciera ser que son los paises mas felices de esta encuesta, Europa occidental muesta un bigote mas largo y se acerca mucho al valor 5 de la puntuacion de felicidad, lo que indica que muchos de estos paises no tiene una felicidad tan alta. Australia y Nueva Zelanda se indican como las zonas mas felices de esta encuesta.
Latinoamerica y el caribe se imponen tambien con puntajes altos de felicidad, la mediana de la caja de latinoamerica y el Caribe se encuentra sobre 6, sin embargo la caja tiene una desviacion a la derecha lo que indica que la mayoria de los paises de latinoamerica y el caribe tienen una felicidad alta, sin embargo en este boxplot nos encontramos con un valor atipico o outliers, este valor se podria explicar talvez por algun pais que pueda estar pasando por algun proceso delicado que implique la poblacion de esa zona se muestre preocupada..
Los paises con menor puntaje de felicidad se ecuentran en las zonas de Africa y Sureste Asiatico, esto se podrioa explicar por lo niveles de pobreza y falta de opotunidades de dsarrollo de alguno de esos paises de esas regiones, lo que indica que esas regiones enfrentan mayores desafios en terminos de bienestar.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Happiness Score', y='Health (Life Expectancy)', hue='Region', data=df, color='skyblue')
plt.title("Boxplot de Puntuacion de felicidad por region")
plt.xlabel("Puntuacion de felicidad")
plt.ylabel("Region")
plt.show()
```
